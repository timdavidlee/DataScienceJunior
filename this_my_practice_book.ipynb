````python
obj_a = dataframe
obj_b = summarized

merged_dataframe = pd.merge(obj_a,
                            obj_b,
                            how='left',
                            left_on=cols_to_group,
                            right_on=cols_to_group )
````

### How to pivot

First, summarize, and condense the rows

````python
cols_to_group = ['MSZoning', 'Street']
value_col = 'LotArea'
summarized = dataframe.groupby(cols_to_group)[value_col].mean().reset_index()
summarized.head()
````

then pivot the summarized data for a more useful format

````python
summarized.pivot_table(values=value_col, index='MSZoning', columns='Street')
````

### can swap around some fields for a different format

````python
summarized.pivot_table(values=value_col, index='Street', columns='MSZoning')
````

````python
summarized.pivot_table(values=value_col, index='Street', columns='MSZoning').transpose()
````
